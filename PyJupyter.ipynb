```python
import ipywidgets as widgets
from ipywidgets import interact
from IPython.display import display, Markdown

def show_table(val1, option, val2):
    num1 = val1[0]
    num2 = val2[0]
    
    if option == "+":
        res = num1 + num2
    elif option == "-":
        res = num1 - num2
    elif option == "*":
        res = num1 * num2
    elif option == "/":
        res = num1 / num2 if num2 != 0 else "Error"
    
    table = f"""
        | num1 | num2 | Result |
        | :---: | :---: | :---: |
        | {num1} | {num2} | {res} |
        """
    display(Markdown(table))

interact(
    show_table,
    val1=widgets.SelectionRangeSlider(description='Num 1', options=range(10), index=(0, 1)),
    option=widgets.Dropdown(description='Operator', options=['*', '/', '+', '-'], value='+'),
    val2=widgets.SelectionRangeSlider(description='Num 2', options=range(10), index=(0, 1)),
)
    
```

```python
import ipywidgets as widgets
from ipywidgets import interact
from IPython.display import display, Markdown

def show_text(text):
    mistake=["Вова", "Дима"]
    for i in mistake:
        text = text.replace(i, "***")
    
    display(Markdown(text))

interact(
    show_text,
    text=widgets.Textarea(description='Textarea:', value='Строка'),
)

```

```python
import ipywidgets as widgets
from IPython.display import display, Markdown

def deisplay_widgets(title, widgets_arr):
    display(Markdown("# " + title))
            
    for i in range(len(widgets_arr)):
        display(widgets_arr[i])

number_widgets = [
    widgets.IntSlider(description='IntSlider:',value=5, min=0, max=10),
    widgets.FloatSlider(description='FloatSlider:', value=0.5, min=0, max=1, step=0.1),
    widgets.IntText(description='IntText:', value=10),
    widgets.FloatText(description='FloatText:', value=3.14),
    widgets.BoundedIntText(description='BoundedIntText:', value=5, min=0, max=10),
    widgets.BoundedFloatText(description='BoundedFloatText:', value=0.5, min=0, max=1),
    widgets.IntProgress(description='IntProgress:', value=6, min=0, max=10),
    widgets.FloatProgress(description='FloatProgress:', value=0.4, min=0, max=1)
]

text_widgets = [
    widgets.Label(value='Я просто текст'),
    widgets.Text(description='Text:', value='Привет'),
    widgets.Textarea(description='Textarea:', value='Строка\nВторая строка'),
    widgets.HTML(description='HTML:', value='<b>Жирный текст</b>'),
    widgets.HTMLMath(description='HTMLMath:', value='x^2 + y^2 = z^2')
]

bool_widgets = [
    widgets.Checkbox(description='Checkbox', value=True),
    widgets.ToggleButton(description='ToggleButton', value=False),
    widgets.ToggleButtons(description='ToggleButtons', options=['A','B','C'], value='A')
]

lists_widgets = [
    widgets.Dropdown(description='Dropdown', options=['Красный','Синий'], value='Красный'),
    widgets.SelectionSlider(description='SelectionSlider', options=['Низ','Сред','Выс'], value='Сред'),
    widgets.SelectionRangeSlider(description='SelectionRangeSlider', options=range(10), index=(2,5)),
    widgets.RadioButtons(description='RadioButtons', options=['Да','Нет'], value='Да'),
    widgets.Select(description='Select', options=['A','B','C']),
    widgets.SelectMultiple(description='SelectMultiple', options=['A','B','C'])
]

buttons_widgets = [
    widgets.Button(description='Button'),
    widgets.Accordion(description='Accordion', children=[widgets.Text(value='Привет')]),
    widgets.Tab(description='Tab', children=[widgets.Text(), widgets.IntSlider()])
]

other_widgets = [
    widgets.HBox([widgets.IntSlider(), widgets.Button()]),
    widgets.ColorPicker(value='red', description='Цвет')
]


deisplay_widgets("Number Widgets", number_widgets)
deisplay_widgets("Text Widgets", text_widgets)
deisplay_widgets("Bool Widgets", bool_widgets)
deisplay_widgets("Lists Widgets", lists_widgets)
deisplay_widgets("Buttons Widgets", buttons_widgets)
deisplay_widgets("Others Widgets", other_widgets)

```
